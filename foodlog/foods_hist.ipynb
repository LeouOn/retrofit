```python
import sys
import numpy as np
from pandas import Series
import matplotlib.pyplot as plt
%matplotlib inline

foods_filename = "./Foods.csv"
foods = []
```

```python
try:
    with open(foods_filename, 'r') as f:
        foods = [food_row.strip() for food_row in f]
except IOError as e:
   print ("I/O error({}): {}".format(e.errno, e.strerror))
except: #handle other exceptions
   print ("Unexpected error:", sys.exc_info()[0])
```

```python
food_series = Series(foods)
hist = food_series.value_counts()
hist[hist > 3]
```

```python
hist[hist > 3].plot(kind='bar')
```
